```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\Prime\\Downloads\\Product_Sales.csv")
df
```

```python
print(plt.style.available)
```

```python
#Q1) Create a line plot showing the trend of sales over the months. Add appropriate labels and a title.

plt.style.use("seaborn-v0_8")
plt.plot(df["Month"],df["Sales"],marker="o",color="Purple")
plt.title("Sales_Over_Months")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
#Q2) Create a pie chart showing the distribution of customers across different months.
value=[0,0.2,0,0.2,0,0.2,0]

plt.style.use("dark_background")
plt.pie(df["Customers"],labels=df["Month"],autopct="%1.1f%%",explode=value,shadow=True)
plt.title("Customers_over_Months")
plt.show()
```

```python
#Q3) Create a stacked bar chart showing both sales and profit for each month. Use different colors for sales and profit.
plt.style.use("dark_background")
plt.grid()
plt.bar(df["Month"],df["Sales"],color="white",linewidth=0.5,edgecolor="white",alpha=0.9,label="Sales")
plt.title("Sales & Profit of Month")
plt.bar(df["Month"],df["Profit"],color="red",linewidth=0.5,edgecolor="white",alpha=0.9,label="Profit")
plt.xlabel("Month")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
#Q4) Create a scatter plot to visualize the relationship between sales and profit. Use Sales on the x-axis and Profit on the y-axis.  

plt.style.use("Solarize_Light2")
plt.scatter(df["Sales"],df["Profit"],marker="*",color="red",s=100)
plt.grid()
plt.title("Sales & Profit",color="black")
plt.xlabel("Sales",color="black")
plt.ylabel("Profit",color="black")
plt.show()
```

```python
#Q5) Plot a multi-line chart to show both sales and profit over the months. Use Month on the x-axis, and Sales and Profit on the y-axis. 

plt.style.use("seaborn-v0_8")
plt.plot(df["Month"],df["Sales"],marker="o",color="Purple",label="Sales")
plt.plot(df["Month"],df["Profit"],marker="o",color="green",label="Profit")
plt.legend()
plt.title("Sales_Over_Months")
plt.xlabel("Months")
plt.ylabel("Sales & Profit")
plt.grid()
plt.show()
```

```python
plt.style.use("seaborn-v0_8")
plt.fill_between(df["Month"],df["Sales"],color="Purple",label="Sales")
plt.fill_between(df["Month"],df["Profit"],color="green",label="Profit")
plt.legend(loc="upper left")
plt.title("Sales_Over_Months")
plt.xlabel("Months")
plt.ylabel("Sales & Profit")
plt.grid()
plt.show()
```
