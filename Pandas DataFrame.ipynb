```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\Prime\\Downloads\\titanic_dataset.csv")

df
```

```python
df.isna().sum()
```

```python
df.drop(columns=["Cabin"],inplace=True)

df
```

```python
df["Age"]=df["Age"].fillna(df["Age"].median())
```

```python
df["Fare"]=df["Fare"].fillna(df["Fare"].median())
```

```python
df.isna().sum()
```

```python
df.rename(columns={"Pclass":"Passanger_class","Parch":"Parent_childern","Embarked":"Port"},inplace=True)
```

```python
df.head(2)
```

```python
def age_category(x):
    if x<18:
        return "Childern"
    elif x<50:
        return "Adult"
    else:
        return "old"
```

```python
df["age_category"]=df["Age"].apply(age_category)
```

```python
df.head(5)
```

```python
pivot_table=df.pivot_table(values="Fare",index="Sex",columns="Passanger_class",aggfunc="mean")

pivot_table
```

```python
survived=df.groupby(["age_category"])["Survived"].sum().sort_values(ascending=False).reset_index()

survived
```
