```python
#Making a list of missing value types

missing_values=["na","--"]

df=pd.read_csv("C:\\Users\\Prime\\Downloads\\bng_property.csv",na_values=missing_values)
df
```

```python
#exception handling

count=0   #counter

for row in df ["OWN_OCCUPIED"]:    #membership operator
    try:
        int(row)
        df.loc[count,"OWN_OCCUPIED"]=np.nan
    except ValueError:
        pass
    count=count+1
```

```python
df
```

```python
count=0

for row in df["NUM_BATH"]:
    try:
        float(row)
        pass
    except ValueError:
        df.loc[count,"NUM_BATH"]=np.nan
    count+=1

df
```

```python
df['PID']=df['PID'].interpolate()
df
```

```python
df['ST_NUM']=df['ST_NUM'].interpolate()
df
```

```python
#using value_counts we define more frequency of values or counts

df['OWN_OCCUPIED']=df['OWN_OCCUPIED'].fillna('Y')

df
```

```python
df.dtypes
df['NUM_BATH']=df['NUM_BATH'].astype('float')

df.dtypes
```

```python
df['NUM_BATH']=df['NUM_BATH'].fillna(df.NUM_BATH.median())
df
```

```python
df['NUM_BEDROOMS']=df['NUM_BEDROOMS'].fillna(df.NUM_BEDROOMS.median())
df
```

```python
df['SQ_FT']=df['SQ_FT'].fillna(df.SQ_FT.mean())
df
```

```python
df['NUM_BATH']=df['NUM_BATH'].astype(int)
df['NUM_BEDROOMS']=df['NUM_BEDROOMS'].astype(int)
df['PID']=df['PID'].astype(int)
df['SQ_FT']=df['SQ_FT'].astype(int)
df['ST_NUM']=df['ST_NUM'].astype(int)

df
```
