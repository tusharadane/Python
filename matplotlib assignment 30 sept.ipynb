```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
```

```python
std=pd.read_csv("C:\\Users\\Prime\\Downloads\\.ipynb_checkpoints\\Student_Dataset.csv")
std
```

```python
#Create a bar chart showing the average score of each student across all subjects.
std["average_score"]=std[["Math","Science","English","History"]].sum(axis=1)/4

std
```

```python
plt.bar(std["Name"],std["average_score"],color="red",linewidth=1,edgecolor="black",alpha=0.5)
plt.title("Average_Score_Of_Student",alpha=0.9)
plt.ylabel("Average_Score")
plt.xticks(rotation=45)
plt.show()
```

```python
#Create a pie chart representing the distribution of scores in Math for all students.
value=[0,0.2,0,0.2,0,0.2,0,0.2,0,0.2]

plt.pie(std["Math"],labels=std["Name"],autopct="%1.1f%%",shadow=True,explode=value)

plt.show()
```

```python
#Create a line chart showing the trend of scores in Science for each student.

plt.plot(std["Name"],std["Science"],color="blue")
plt.title("Science_Score_of_Student")
plt.ylabel("Marks")
plt.xlabel("Student_Name")
plt.xticks(rotation=45)
plt.show()
```

```python
#Create a scatter plot comparing the scores of Math and English for each student.

plt.scatter(std["Math"],std["English"],marker="*",s=100,color="r")
plt.grid(linestyle="--",alpha=0.4)
plt.xlabel("English_Sub_Marks")
plt.ylabel("Math_Sub_Marks")
plt.show()
```
