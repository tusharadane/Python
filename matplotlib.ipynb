```python
df=pd.read_excel("C:\\Users\\Prime\\Downloads\\Sample - Superstore.xlsx")
df.head()
```

```python
df["order_year"]=df["Order Date"].dt.year
```

```python
df.head()
```

```python
group_year=df.groupby("order_year")["Sales"].sum().reset_index()
group_year
```

```python
plt.plot(group_year["order_year"],group_year["Sales"],marker="o")
plt.title("Year_wise_Sales")
plt.xlabel("Sales")
plt.ylabel("Years")
plt.show()
```

```python
df["quarter"]=df["Order Date"].dt.to_period("Q")

quarter_wise_sales=df.groupby("quarter")[["Sales","Profit"]].sum().reset_index()
quarter_wise_sales
plt.figure(figsize=(12,8))
plt.style.use("dark_background")
plt.plot(quarter_wise_sales["quarter"].astype(str),quarter_wise_sales["Sales"],marker="o",color="g",label="Sales")
plt.plot(quarter_wise_sales["quarter"].astype(str),quarter_wise_sales["Profit"],marker="o",color="orange",linestyle="--",label="Profit")

plt.title("Quarter Wise Sales & Profit")
plt.xlabel("Quarter")
plt.ylabel("Values")

plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
df["quarter"]=df["Order Date"].dt.to_period("Q")

quarter_wise_sales=df.groupby("quarter")[["Sales","Profit"]].sum().reset_index()
quarter_wise_sales
plt.figure(figsize=(12,8))
plt.style.use("Solarize_Light2")
plt.fill_between(quarter_wise_sales["quarter"].astype(str),quarter_wise_sales["Sales"],color="g",label="Sales")
plt.fill_between(quarter_wise_sales["quarter"].astype(str),quarter_wise_sales["Profit"],color="orange",linestyle="--",label="Profit")

plt.title("Quarter Wise Sales & Profit")
plt.xlabel("Quarter")
plt.ylabel("Values")

plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
correlation_matrix=df.corr(numeric_only=True)
correlation_matrix
```

```python
plt.imshow(correlation_matrix,cmap="coolwarm")
plt.title("Corelation")
plt.colorbar()
plt.show()
```
