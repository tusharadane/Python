```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
df = pd.read_excel("C:\\Users\\LENOVO\\Documents\\Microsoft Excel\\Sample - Superstore.xlsx")
df.head()
```

```python
df["Order Year"] = df["Order Date"].dt.year
```

```python
df.head(2)
```

```python
grouped_year = df.groupby("Order Year")["Sales"].sum().reset_index()
grouped_year

plt.figure(figsize = (10,8))
plt.plot(grouped_year["Order Year"],grouped_year["Sales"],marker = "o")

plt.title("Year Wise Sales")
plt.xlabel("Years")
plt.ylabel("Sales")

plt.show()
```

```python
df["Quarter"] = df["Order Date"].dt.to_period("Q")
df.head()
```

```python
df.info()
```

```python
!jupyter --version
```

```python
plt.style.use("dark_background")

Quarter_Wise_Sales = df.groupby("Quarter")[["Sales","Profit"]].sum().reset_index()
Quarter_Wise_Sales

plt.figure(figsize=(12,8))
plt.plot(Quarter_Wise_Sales["Quarter"].astype(str),Quarter_Wise_Sales["Sales"],marker="o",color = "g",label="Sales")

plt.plot(Quarter_Wise_Sales["Quarter"].astype(str),Quarter_Wise_Sales["Profit"],marker="o",linestyle = "--",color = "orange",label = "Profit")

plt.title("Quarter Wise Sales and Profit")
plt.xlabel("Quarter")
plt.ylabel("Values")


plt.xticks(rotation = 45)


plt.legend()

plt.tight_layout()
plt.grid(False)
plt.show()
```

```python
# Creating a sample dataset
np.random.seed(42)
data = {
    'Age': np.random.randint(18, 65, 50),
    'Income': np.random.randint(30000, 120000, 50),  # Income in dollars
    'Years_of_Education': np.random.randint(10, 20, 50),
    'Work_Experience': np.random.randint(1, 40, 50),
    'Health_Score': np.random.uniform(50, 100, 50),  # Health score out of 100
    'Loan_Amount': np.random.randint(5000, 50000, 50),
    'Credit_Score': np.random.randint(300, 850, 50),  # Credit score range
    'Satisfaction_Score': np.random.randint(1, 10, 50)  # Customer satisfaction (1-10)
}

df = pd.DataFrame(data)


df.head()
```

```python
correlation_matrix = df.corr(numeric_only = True)
correlation_matrix 
```

```python
cmap_names = plt.colormaps()
print(cmap_names)
```

```python
plt.figure(figsize = (12,8))

plt.imshow(correlation_matrix ,cmap = "viridis")

plt.colorbar()

plt.xticks(range(len(correlation_matrix)),correlation_matrix.columns,rotation = 45)
plt.yticks(range(len(correlation_matrix)),correlation_matrix.columns)

plt.title("Correlation Matrix")


plt.grid(False)
plt.show()
```

```python
plt.style.use("dark_background")

plt.figure(figsize = (12,8))

plt.imshow(correlation_matrix ,cmap = "viridis")

plt.colorbar()

plt.xticks(range(len(correlation_matrix)),correlation_matrix.columns,rotation = 45)
plt.yticks(range(len(correlation_matrix)),correlation_matrix.columns)

for i in range(len(correlation_matrix)):
    for j in range(len(correlation_matrix)):
        plt.text(j,i,f'{correlation_matrix.iloc[i,j]:.2f}',ha = "center",va = "center",color = "black")

plt.title("Correlation Matrix")

plt.grid(False)
plt.show()
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Create a boxplot with customizations
box = ax.boxplot(df['Income'], 
                 patch_artist=True,  # Allows us to color the box
                 boxprops=dict(facecolor='lightblue', color='blue'),  # Box color
                 whiskerprops=dict(color='blue', linewidth=2),  # Whisker color and width
                 capprops=dict(color='blue', linewidth=2),  # Caps color and width
                 medianprops=dict(color='red', linewidth=2),  # Median line color and width
                 flierprops=dict(marker='o', markerfacecolor='orange', markersize=8, linestyle='none'))  # Outlier customization

# Customize the plot
ax.set_title('Income Distribution', fontsize=16, fontweight='bold', color='darkblue')  # Adding a title
ax.set_ylabel('Income', fontsize=14)  # Label for y-axis
ax.set_xticks([1])  # Custom x-tick
ax.set_xticklabels(['Income'], fontsize=12)  # Label for the x-axis

# Set background grid
ax.grid(False)

# Show the plot
plt.tight_layout()
plt.show()
```
