```python
#Which bowler has taken the most wickets?

ipl_data.groupby("bowler")["is_wicket"].sum().sort_values(ascending=False).head(3)
```

```python
#What is the highest partnership between two batsmen in an inning?

ipl_data.groupby(["batsman","non_striker"])["batsman_runs"].count().sort_values(ascending=False).head(3)
```

```python
#Which team has conceded the most extras (wides, no balls, etc.) in a season?

ipl_data.groupby("bowling_team")["extra_runs"].count().sort_values(ascending=False).head(3)
```
