```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/NeilPandey/Test_Dataset/refs/heads/master/diabetes.csv')

df.head()
```

```python
feature_col=['Pregnancies','Glucose','BloodPressure','Insulin','BMI','DiabetesPedigreeFunction','Age']
```

```python
x=df[feature_col] #df.drop[outcome]

#y=df.[outcome]
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)

x_test.head()
```

```python
from sklearn.linear_model import LogisticRegression

logreg=LogisticRegression()

logreg.fit(x_train,y_train)
```

```python
y_pred=logreg.predict(x_test)

res=sum(y_pred!=y_test)

res
```

```python
from sklearn import metrics

cnf_matrix=metrics.confusion_matrix(y_test,y_pred)

cnf_matrix
```

```python
print("accuracy:",metrics.accuracy_score(y_test,y_pred))
```
